## ДЗ 7. Нестандартные ситуации

На основе выбранных продуктов из первого вебинара (на примере Telegram), выберите 2 A/B-теста и составьте план эксперимента, чтобы избежать Peeking Problem
Распишите, какие метрики и как вы будете анализировать

### A/B-тест №1: Изменение дизайна кнопки "Отправить" в приложении Telegram

Гипотеза: Изменение дизайна кнопки "Отправить" на более яркую и заметную привлечет больше внимания пользователей и увеличит частоту отправки сообщений.

Разделение группы: Пользователи приложения Telegram, случайным образом разделенные на две группы: контрольную (A) и экспериментальную (B).

Описание эксперимента:
- Вариант А: Контрольная группа, которая видит стандартный дизайн кнопки "Отправить".
- Вариант В: Экспериментальная группа, которая видит измененный дизайн кнопки "Отправить" с более выразительными цветами и формой.

Метрики:
1. Конверсия отправки сообщений: Измеряется как процент пользователей, видевших измененный дизайн кнопки "Отправить" и совершивших отправку сообщения в приложении Telegram.
2. Время нахождения в приложении: Измеряется как среднее время, проведенное пользователями в приложении Telegram после применения измененного дизайна кнопки "Отправить".

Анализ:
Для избежания Peeking Problem (эффекта подглядывания) необходимо использовать randomization unit, то есть случайно назначить каждому уникальному пользователю одну из двух групп. Анализ проводится путем сравнения результатов метрик между контрольной и экспериментальной группами с использованием статистических тестов, таких как t-тест.


### A/B-тест №2: Размещение кнопки "Поделиться" на экране чата в приложении Telegram

Гипотеза: Перемещение кнопки "Поделиться" на более видное и удобное место на экране чата повысит активность пользователей в использовании функции "Поделиться" контентом.

Разделение группы: Пользователи приложения Telegram, случайным образом разделенные на две группы: контрольную (A) и экспериментальную (B).

Описание эксперимента:
- Вариант А: Контрольная группа, которая видит стандартное размещение кнопки "Поделиться" на экране чата.
- Вариант В: Экспериментальная группа, которая видит измененное размещение кнопки "Поделиться" на более видном месте на экране чата.

Метрики:
1. Частота использования кнопки "Поделиться": Измеряется как количество раз, когда пользователи нажимают на кнопку "Поделиться" в приложении Telegram.
2. Распространение контента: Измеряется как количество пользователей, получивших контент, поделенный с помощью кнопки "Поделиться".

Анализ:
Для избежания эффекта подглядывания необходимо использовать randomization unit, чтобы случайным образом назначить каждому уникальному пользователю одну из двух групп. Анализ проводится путем сравнения результатов метрик между контрольной и экспериментальной группами с использованием статистических тестов, таких как t-тест

Randomization unit (случайная единица) - это единица, на основе которой случайным образом назначаются участники эксперимента в разные группы. Он используется для избежания эффекта подглядывания (peeking problem) и обеспечения случайного распределения участников между контрольной и экспериментальной группами.

Randomization unit может быть различным в зависимости от контекста эксперимента. В случае приложений и веб-сайтов это часто является уникальным идентификатором пользователя или сессии. Например, каждый пользователь или сессия может быть случайным образом назначен в одну из двух групп эксперимента: контрольную и экспериментальную.

Когда случайное назначение происходит на уровне randomization unit, это гарантирует, что каждый уникальный пользователь или сессия имеет одно и тоже самое условие эксперимента на протяжении всего периода эксперимента. Это позволяет сравнить результаты метрик между группами с высокой степенью доверия и исключить возможные искажения, вызванные несбалансированным случайным распределением участников.

Таким образом, randomization unit обеспечивает рандомизацию участников эксперимента, аналогично подбрасыванию монеты или броску кубика, чтобы гарантировать случайное и сбалансированное распределение участников между группами контроля и эксперимента. Это помогает создать надежную основу для анализа результатов эксперимента и сделать выводы о влиянии изменений на метрики и поведение пользователей.
